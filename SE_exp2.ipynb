```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def quadratic(x,a=-1,b=0,c=10):
  return a*x**2+b*x+c
```

```python
stages = ["Requirements","Design","Implementation","Testing","Deployment","Maintenance"]
x_pos = np.linspace(-2,2,len(stages))
y_pos = quadratic(x_pos)
#plot the quadratic curve
x_curve = np.linspace(-2.5,2.5,500)
y_curve = quadratic(x_curve)
plt.figure(figsize=(10,6))
plt.plot(x_curve,y_curve,color = "blue", linestyle = "--",label= "Quadratic curve")
#adding makers and connect them with vertical lines to create the waterfall effect
for i in range(len(stages)):
  if i>0:
    #draw vertical drop
    plt.plot(#horizontal segment
        [x_pos[i-1],x_pos[i]],
        [y_pos[i-1],y_pos[i-1]],
        color = 'blue',
        linewidth = 2
    )
    plt.plot(#vertical drop
        [x_pos[i],x_pos[i]],
        [y_pos[i-1],y_pos[i]],
        color = 'blue',
        linewidth = 2
    )
    plt.scatter(x_pos[i],y_pos[i],color = "red",zorder=5)
    plt.text(x_pos[i],y_pos[i]+0.3,stages[i],ha="center",fontsize=10)
plt.title("Waterfall Model representation using quadratic curve",fontsize=10)
plt.xlabel("Stages",fontsize = 12)
plt.ylabel("Value",fontsize=12)
plt.axhline(0,color="black",linewidth=0.5,linestyle="--")
plt.axvline(0,color="black",linewidth=0.5,linestyle="--")
plt.grid(color="gray",linestyle="--",linewidth=0.5,alpha=0.7)
plt.legend()
plt.show()
```
